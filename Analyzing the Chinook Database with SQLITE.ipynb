## Finding Best Genres for USA Market 

A new records comnpany wants to get in the American market. For advertising purposes, they want to know which are the top 3 best selling jenres in the USA.

```python
%%sql

WITH usa_sold AS
   (
    SELECT il.* FROM invoice_line il
    INNER JOIN invoice i on il.invoice_id = i.invoice_id
    INNER JOIN customer c on i.customer_id = c.customer_id
    WHERE c.country = "USA"
   )
SELECT 
    g.name,
    count(invoice_id) as total_invoices,
    round(
        (cast(count(susa.invoice_id) AS FLOAT) / (
        SELECT COUNT(*) 
        FROM usa_sold ) * 100)
        , 2)
     percentages_sold
FROM usa_sold susa
INNER JOIN track t on t.track_id = susa.track_id
INNER JOIN genre g on g.genre_id = t.genre_id
GROUP BY g.name
ORDER BY total_invoices desc
LIMIT 3;
```

Based on the findings above, the top 3 best sellers genres in the USA are Rock, Alternative & Punk and Metal.
The Rock genre is in the lead with 53% of all sales.


## Analyze Employees Sales Performance

```python
%%sql

WITH employee_info AS (
SELECT
    e.employee_id,
    e.first_name || ' ' || e.last_name AS employee_name,
    e.title,
    e.reports_to,
    e.hire_date,
    il.invoice_id,
    il.unit_price,
    il.quantity,
    i.total
FROM employee e
LEFT JOIN customer c ON c.support_rep_id = e.employee_id
LEFT JOIN invoice i ON i.customer_id = c.customer_id
LEFT JOIN invoice_line il ON il.invoice_line_id = i.invoice_id
    )

SELECT 
    employee_name,
    round(sum(total), 2) as total_sales
FROM employee_info 
GROUP BY employee_id 
ORDER BY total_sales DESC
LIMIT 3;
```

## Analyze Sales by Country

Next, we will analyize the sales by countries.
All countries with a single customer will be grouped into a united group.

```python
#count number of customers per country
%%sql
 WITH countries AS (
    SELECT 
        c.country,
        count(c.customer_id) total_customers
    FROM customer c
    GROUP BY c.country
    ORDER BY total_customers desc
 )
SELECT *
FROM countries
     
```

Group countries with one customer into 'Other', and find fore each country:
1 - The total number of customers 
2 - The total sales value 
3 - The average order value 
4 - The average order value 

```python
%%sql
WITH countries_grouped as (
    SELECT 
        CASE WHEN COUNTRY IN(
                SELECT country FROM (
                SELECT country, count(*) as total
                FROM customer
                GROUP BY country )
                where total = 1) then 'Other'
        ELSE c.country
        END AS countries,
        customer_id
    FROM customer c 
)

SELECT 
    countries,
    total_customers,
    total_sales,
    avg_order_value,
    customer_lifetime_value
    
FROM (
    SELECT
        countries, 
        count(distinct cg.customer_id) total_customers,
        round(sum(i.total) ,2) total_sales,
        round(sum(il.unit_price) / count(distinct i.invoice_id) ,2) avg_order_value,
        round(sum(il.unit_price) / count(distinct cg.customer_id) ,2) customer_lifetime_value,
            CASE 
                WHEN countries = 'Other' THEN 1
                ELSE 0
                END AS sort_other
    FROM countries_grouped cg
    JOIN invoice i ON i.customer_id = cg.customer_id
    JOIN invoice_line il ON il.invoice_id = i.invoice_id
    GROUP BY countries
    
)

ORDER BY sort_other, total_sales desc




    
   
```
